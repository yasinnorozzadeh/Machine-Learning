```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
X = np.random.uniform(0, 90, 250)
Y = abs(X + np.random.uniform(-10, 10, 250) * np.random.uniform(1, 3, 250))
slope, intercept, r_value, p_value, std_err = stats.linregress(X,Y)

plt.scatter(X, Y)
```

```python
plt.scatter(X, Y)
plt.plot(X, (X * 0.8), c='g', lw=4, marker=",")
plt.show()
```

```python
X = X.reshape((250, 1))
Y = Y.reshape((250, 1))
```

```python
plt.scatter(X, Y)
plt.plot(X, X * np.matmul(np.linalg.inv(np.matmul(X.T, X)), np.matmul(X.T, Y)), c="#4CAF50", lw=4, marker="_")
plt.show()
```

```python
plt.plot(X ,Y, 'o', label='original data')
plt.plot(X ,X * np.matmul(np.linalg.inv(np.matmul(X.T, X)), np.matmul(X.T, Y)), c= 'orange',lw=4.5,alpha=1)
plt.plot(X, intercept + slope * X, 'r' ,linestyle = 'dotted',lw=4, alpha=0.6)

plt.legend()
plt.show()
```
